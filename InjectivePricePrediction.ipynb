# INJECTIVE PRICE PREDICTION (2022)

```python
# Import Library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Baca Dataset
df = pd.read_csv('INJ.csv')
```

```python
# Preview Dataset
print(df.head())
```

```python
# Preprocessing data
print("Missing values:\n", df.isnull().sum())
```

```python
# Hapus bari yang memiliki nilai Nan
df.dropna(inplace=True)
```

```python
# Ubah kolom 'date' menjadi format datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Urutkan data berdasarkan tanggal
df.sort_values('date', inplace=True)
```

```python
# Buat label untuk klasifikasi (1 jika naik, 0 jika turun)
df['price_direction'] = np.where(df['close'].shift(-1) > df['close'], 1, 0)
```

```python
# Gunakan fitur yang relevan, sepert open, high, low
features = ['open', 'high', 'low']
x = df[features]
y = df['price_direction']
```

```python
# Membagu data menjadi data latih dan data uji
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2, random_state=42)
```

```python
# Buat model menggunakan data latih
model = DecisionTreeClassifier(random_state=42)
```

```python
# Latih model menggunakan data latih
model.fit(x_train, y_train)
```

```python
# Prediksi menggunakan data uji
y_pred = model.predict(x_test)
```

```python
# Hitung akurasi model
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi model: {accuracy * 100:2f}%')

# Prediksi arah harga untuk hari berikutnya menggunakan data terakhir
next_day_features = x.iloc[-1].values.reshape(1, -1)
next_day_prediction = model.predict(next_day_features)

# Tampilkan hasil prediksi
if next_day_prediction[0] == 1:
    print("Model memprediksi bahwa harga INJECTIVE akan NAIK besok")
else:
    print("Model memprediksi bahwa harga INJECTIVE akan TURUN besok")
```
