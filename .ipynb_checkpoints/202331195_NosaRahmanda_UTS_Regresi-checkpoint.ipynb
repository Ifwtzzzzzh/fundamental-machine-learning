```python
# IMPORT LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# MEMUAT DAN MENAMPILKAN DATASET
df = pd.read_csv("R01_students_performance.csv")
print(df.head())
print(df.info())
```

```python
# VISUALISASI AWAL KORELASI STUDY HOUR DAN EXAM SCORE
plt.figure(figsize=(8,6))
sns.scatterplot(x="Study Hours (X)", y="Exam Scores (Y)", data=df)
plt.title("Relationship between Study Hour and Exam Score")
plt.xlabel("Study Hour")
plt.ylabel("Exam Score")
plt.show()
```

```python
# MENDEFINISIKAN X DAN Y
X = df[["Study Hours (X)"]]
y = df["Exam Scores (Y)"]
```

```python
# INISIALISASI DAN MELATIH MODEL REGRESI
model = LinearRegression()
model.fit(X, y)
```

```python
# MEMBUAT KOEFISIEN DAN INTERCEPT
intercept = model.intercept_
coefficient = model.coef_[0]
print(f"Persamaan regresi: Exam Score = {intercept:.2f} + {coefficient:.2f} * Study Hour")
```

```python
# MENGHITUNG R SQUARED DAN MEAN SQUARRED ERROR
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
print(f"R-squared: {r2:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
```

```python
# VISUALISASI GARIS REGRESI
plt.figure(figsize=(8,6))
sns.scatterplot(x="Study Hours (X)", y="Exam Scores (Y)", data=df, label="Data")
plt.plot(X, y_pred, color="red", label="Regression Line")
plt.title("Regression of Study Hour vs Exam Score")
plt.xlabel("Study Hour")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```
